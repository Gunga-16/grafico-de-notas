# Import



```python
# Importando a bibiblioteca que gera gráficos
import matplotlib.pyplot as plt
```

# Variáveis

```python
# Variáveis sobre o aluno como por exemplo: Primeiro Nome, Número da Chamada e Turma.
aluno = 'Sérgio'
numero_da_chamada = str(43)
turma = '2°G'


# Variável lista das 14 matérias
materias = [
    " Biot. ",
    " Soc. ",
    " Geo. ",
    " Rob. ",
    " Ing. ",
    " Edu. ",
    " Mat. ",
    " His. ",
    " Pro. ",
    " Por. ",
    " Emp. ",
    " Quí. ",
    " Proj. ",
    " Bio. "
]



notas = [
    10.0,
    10.0,
    10.0,
    10.0,
    10.0,
    9.8,
    9.8,
    9.7,
    9.6,
    9.6,
    9.0,
    9.0,
    9.0,
    8.0
]
```

# Função

```python
# Função para gerar o gráfico.

def grafico_de_notas():
 x = materias
 y = notas
 plt.plot(x, y, marker='o')
 plt.xlabel('Matérias')
 plt.ylabel('Notas')
 plt.title(aluno + ' - ' + numero_da_chamada + ' - ' + turma)
```

```python
# Chamando a função "grafico_de_notas".

grafico_de_notas()
```
